## Gráfico simple

Se hará un gráfico de la curva $y=x^{2}$

```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y)
# para poner un título al gráfico
plt.title('y=x^2')
# para poner etiqueta al eje x
plt.xlabel('x')
# para poner etiqueta al eje y
plt.ylabel('y')
# para mostrar el gráfico desde un script de python tendría 
# que llamar plt.show(), pero no es necesario porque estamos en jupyter
```

## Múltiples curvas en un solo gráfico

Se presentarán en un solo gráfico las curvas de $y=x^{2}$, $y=x^{3}$ y $y=x^{4}$

```python
y3 = x**3
y4 = x**4
# el gráfico ahora incluye leyenda
plt.plot(x,y, label='x^2')
plt.plot(x,y3, label='x^3')
plt.plot(x,y4, label='x^4')
# para mostrar la leyenda
plt.legend()
```

## Múltiples gráficos

```python
# el primer argumento indica la cantidad de filas, el segundo la cantidad de gráficos de la fila, el tercero la posición del gráfico
plt.subplot(1,3,1)
plt.plot(x,y,'r')
plt.subplot(1,3,2)
plt.plot(x,y3,'g')
plt.subplot(1,3,3)
plt.plot(x,y4,'b')
# para evitar que los gráficos estén muy pegados
plt.tight_layout()
```

## Manipulación de figuras

```python
# crea una figura con el ancho y el alto (en ese orden) indicado en pulgadas por la opción figsize
fig = plt.figure(figsize=(5,4))
# caracteristicas de los ejes (en orden) izquierda, parte baja, ancho y alto en decimales
ejes = fig.add_axes([0.1,0.1,0.9,0.9])
# etiqueta del eje x
ejes.set_xlabel('x')
# etiqueta del eje y
ejes.set_ylabel('y')
# título
ejes.set_title('Potencias')
# gráfico 1
ejes.plot(x,y,label='y=x^2')
# gráfico 2
ejes.plot(y,x,label='y=x^1/2')
# mostrar leyendas
ejes.legend()
```

## Subplots

```python
# dos filas y 3 columnas de subplots con las dimensiones en pulgadas indicadas en figsize
fig, ejes = plt.subplots(figsize=(8,4),nrows=2,ncols=3)
# el segundo subplot de la segunda fila
ejes[1,1].set_title('Plot 2')
ejes[1,1].set_xlabel('x')
ejes[1,1].set_ylabel('y')
ejes[1,1].plot(x,np.sin(x))
# para que no estén pegados
plt.tight_layout()
```

## Opciones de apariencia

```python
# crear una figura con dimensiones (5,6)
fig = plt.figure(figsize=(5,6))
# crear ejes que abarquen todo el espacio
ejes = fig.add_axes([0,0,1,1])
# para elegir un color ver: https://en.wikipedia.org/wiki/Web_colors
color = 'b'
# opacidad
alpha = 0.75
# ancho de linea
lw = 1
# estilo de linea
ls = '--'
# marcador
marker = 'o'
# tamaño del marcador
markersize = 9
# color del marcador
markercolor = 'r'
# para tener una grilla
ejes.grid(True)
# grafico
ejes.plot(
    x,
    y,
    color=color,
    alpha=alpha,
    lw=lw,
    ls=ls,
    marker=marker,
    markersize=markersize,
    markerfacecolor=markercolor
)
```

## Gráficos con DataFrames de pandas

```python
zoo = pd.read_csv('zoo.csv',delimiter=',')
```

```python
# reemplazar los datos faltantes con 0s
zoo = zoo.fillna(0)
# los ID de los animales serán el eje x
x = zoo.values[:,1]
# la cantidad de agua que necesitan los animales serán el eje y
y = zoo.values[:,2]
# características de la figura
fig = plt.figure(figsize=(6,4))
ejes = fig.add_axes([0,0,1,1])
ejes.set_ylabel('Cantidad de agua necesitada')
ejes.set_xlabel('ID del animal')
ejes.set_title('Agua vs ID')
ejes.plot(x,y,label=':o')
ejes.legend()
```

## Texto simbólico

```python
x = np.linspace(0.1,0.6,10)
y1 = x**2
y2 = np.sin(x)
y3 = 1/(2*x)
# figura
fig = plt.figure(figsize=(6,5))
# ejes
ejes = fig.add_axes([0,0,1,1])
# texto -> x,y,r'$$'
# también puede agregar letras griegas -> \alpha, \beta
# símbolos matemáticos como \tilde \vec \hat \bar
# operadores de desigualdad \geq \leq \ne
ejes.text(0.5,0.25,r'$y_{1}=x^{2}$')
ejes.text(0.5,0.47,r'$y_{2}=sin(x)$')
ejes.text(0.2,2.5,r'$y_{3}=\frac{1}{2x}$',fontsize=15)
# solo para ver como queda
ejes.text(0.45,4.5,r'$\tilde a$',fontsize=20)
ejes.text(0.45,4,r'$\overline{a}$',fontsize=20)
ejes.text(0.45,3.5,r'$\vec a$',fontsize=20)
ejes.text(0.45,3,r'$\lim_{x \to 2} f(x) =x $',fontsize=20)
ejes.text(0.45,2,r'$\hat a$',fontsize=20)
ejes.text(0.45,1.5,r'$\bar a$',fontsize=20)
# grafico 1
ejes.plot(x,y1)
# grafico 2
ejes.plot(x,y2)
# grafico 3
ejes.plot(x,y3)
```
